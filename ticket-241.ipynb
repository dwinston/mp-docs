How to plot the total energy as a function of volume of unit cell for a relaxed structure, across the Materials Project.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from pymatgen import MPRester
mpr = MPRester()
```

```python
data = mpr.query({},['material_id', 'volume','energy'])
```

```python
x = [d['volume'] for d in data]
y = [d['energy'] for d in data]
plt.plot(x, y, 'bo', alpha=0.05)
plt.xlabel('Volume ($\AA^3$)')
plt.ylabel('Energy (eV)')
```
